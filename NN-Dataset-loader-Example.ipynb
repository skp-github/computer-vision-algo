```python
import torch
import torch.nn as nn 
from torch.optim import SGD
from torch.utils.data import Dataset, DataLoader
```

# Neural Network Class 

```python
class NNetwork(nn.Module):
  def __init__(self):
    super().__init__()
    self.input_to_hidden = nn.Linear(2,8)
    self.activation = nn.ReLU()
    self.hidden_to_output = nn.Linear(8,1)
  
  def forward(self, x):
    x = self.input_to_hidden(x)
    x = self.activation(x)
    x = self.hidden_to_output(x)
    return x
```

#Dataset Class

```python
class MyDataset(Dataset):
  def __init__(self, x, y):
    self.x = torch.tensor(x).float().to('cuda')
    self.y = torch.tensor(y).float().to('cuda')
  def __len__(self):
    return len(self.x)
  def __getitem__(self, ix):
    return self.x[ix], self.y[ix]
```

# Training code

```python
x = [[1,2],[3,4],[5,6],[7,8]]
y = [[3],[7],[11],[15]]
dset = MyDataset(x, y)
dl = DataLoader(dset, batch_size=2, shuffle=True)
mynet = NNetwork().to('cuda')
loss_func = nn.MSELoss()
opt = SGD(mynet.parameters(), lr=0.001)
loss_history = []
for i in range(50):
  for data in dl:
    x, y = data 
    opt.zero_grad()
    loss = loss_func(mynet(x), y)
    loss.backward()
    opt.step()
    loss_history.append(loss)
```
